Figure Anatomy + Subplots (multiple plots in one figure)

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot([1, 2, 3], [1, 4, 9])
ax[1].plot([1, 2, 3], [1, 2, 3])

plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
ax[0,0].plot([1, 2, 3], [1, 4, 9])
ax[0,1].plot([1, 2, 3], [3, 2, 1])
ax[1,0].plot([1, 2, 3], [2, 2, 2])
ax[1,1].plot([1, 2, 3], [1, 8, 27])

plt.show()
```

signature for a 2 x 2 grid -> ax[row, col]

```python
# ML challenge
epochs = [1, 2, 3, 4, 5]
train_loss = [0.9, 0.7, 0.55, 0.42, 0.35]
val_loss = [1, 0.85, 0.70, 0.60, 0.58]

# create 1 x 2 grid
fig, ax = plt.subplots(1, 2)

# left subplot
ax[0].plot(epochs, train_loss)
# right subplot
ax[1].plot(epochs, val_loss)

# show the figure
plt.show()
```
