COLOR + COLORMAPS (HEATMAPS / CONFUSION MATRIX)

```python
# practice

x = [1, 2, 3]
y1 = [1, 4, 9]
y2 = [1, 2, 3]

plt.plot(x, y1, color="green", label="curve")
plt.plot(x, y2, color="orange", label="line")

plt.legend()

plt.show()
```

```python
# color scatter
y_true = [3, 5, 2, 8, 7]
y_pred = [2.5, 5.2, 2.2, 7.6, 6.9]

# get their err
err = list()
for t, p in zip(y_true, y_pred):
    solve = t-p
    solve_n = abs(solve)
    err.append(solve_n)
print(err)

# draw the scatter
plt.scatter(y_true, y_pred, c=err, marker="x")
plt.colorbar()

plt.show()
```

```python
# confusion matrix heatmap
cm = [[8, 2], [1, 9]]

plt.imshow(cm, cmap="viridis")
plt.colorbar()

plt.show()
```

imshow() turns matrix into colors, colorbar() explains the colors which is automatically linked to imshow() and large number - darker color as small number - brighter color for viridis or else "viridis_r" vice versa
