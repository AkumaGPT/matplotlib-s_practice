Figure Anatomy + Subplots (multiple plots in one figure)

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot([1, 2, 3], [1, 4, 9])
ax[1].plot([1, 2, 3], [1, 2, 3])

plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
ax[0,0].plot([1, 2, 3], [1, 4, 9])
ax[0,1].plot([1, 2, 3], [3, 2, 1])
ax[1,0].plot([1, 2, 3], [2, 2, 2])
ax[1,1].plot([1, 2, 3], [1, 8, 27])

plt.show()
```

Signatures: subplots(nrow, ncol), 2 x 2 grid -> ax[row, col]

```python
# ML challenge
epochs = [1, 2, 3, 4, 5]
train_loss = [0.9, 0.7, 0.55, 0.42, 0.35]
val_loss = [1, 0.85, 0.70, 0.60, 0.58]

# create 1 x 2 grid
fig, ax = plt.subplots(1, 2)

# left subplot
ax[0].plot(epochs, train_loss)
# right subplot
ax[1].plot(epochs, val_loss)

# show the figure
plt.show()
```

Common plot types - Line, Scatter, Bar & Histogram

```python
# 1 row 4 plots
fig, ax = plt.subplots(1, 4, figsize=(12, 3))

# for line
epochs = [1, 2, 3, 4, 5]
loss = [0.9, 0.7, 0.55, 0.42, 0.35]
ax[0].plot(epochs, loss)

# for scatter
y_true = [3, 5, 2, 8, 7]
y_pred = [2.5, 5.2, 2.2, 7.6, 6.9]
ax[1].scatter(y_true, y_pred)

# for bar
features = ["age", "income", "score"]
importance = [0.2, 0.6, 0.2]
ax[2].bar(features, importance)

# for histogram
values = [1, 2, 2, 2, 3, 3, 3, 4, 6, 7, 7, 8]
ax[3].hist(values)

plt.show
```

```python
# recall test

import pandas as pd

df = pd.read_csv("salary_dataset.csv")
df.head()
# pick a column, plot a histogram and a bar chart of value_counts().head()
df1 = df["Age"]

fig, ax = plt.subplots(1, 2, figsize=(15, 7))

# for histogram
ax[0].hist(df1.head(600))

# for bar chart

# get the x-axis
age_x = df1.value_counts().head().index.to_list()
age_y = df1.value_counts().head() # y-axis
ax[1].bar(age_x, age_y)

plt.show()
```
